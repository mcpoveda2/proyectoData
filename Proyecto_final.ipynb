# **DESCRIPCION DE LAS VARIABLES**
**id_cliente:** Identificación única del cliente.\
**loan_status:** Estado del préstamo.\
**loan_amnt:** El monto del préstamo solicitado por el prestatario.\
**installment:** La cuota mensual que debe pagar el prestatario si se origina el préstamo.\
**term:** El número de pagos del préstamo. Los valores son en meses y pueden ser 36 o 60.\
**emp_title:** El título del empleo proporcionado por el prestatario al solicitar el préstamo.\
**emp_length:** Duración del empleo en años. Los valores posibles están entre 0 y 10, donde 0 significa menos de un año y 10 significa diez años o más.\
**home_ownership:** Estado de la propiedad de la vivienda proporcionada por el prestatario durante el registro (por ejemplo, RENT, OWN, MORTGAGE, OTHER).\
**annual_inc:** El ingreso anual autodeclarado proporcionado por el prestatario durante el registro.\
**verification_status:** Estado de verificación del ingreso (verificado, no verificado, fuente verificada).\
**purpose:** Categoría proporcionada por el prestatario para la solicitud del préstamo.\
**addr_state:** El estado proporcionado por el prestatario en la solicitud de préstamo.\
**delinq_2yrs:** El número de incidencias de morosidad de 30+ días en el archivo de crédito del prestatario en los últimos 2 años.\
**next_pymnt_d:** Fecha del próximo pago.\
**earliest_cr_line:** La fecha en que se abrió la primera línea de crédito reportada del prestatario.\
**mths_since_last_delinq:** El número de meses desde la última morosidad del prestatario.\
**total_pymnt:** Pago total realizado hasta la fecha.\
**recoveries:** Recuperaciones.\
**collection_recovery_fee:** Comisión de recuperación de cobro.\
**last_pymnt_d:** Fecha del último pago.\
**settlement_status:** Estado de liquidez.\
**application_type:** Indica si el préstamo es una solicitud individual o una solicitud conjunta con dos co-prestatarios.\
**tot_hi_cred_lim:** Límite de crédito más alto.

## Definición de mal pagador

mal_pagador: aquellos clientes que:
1. retraso de mas de 90 dias en sus prestamos actuales
2. que su prestamo haya sido cancelado
3. en los ultimos 2 años se haya retrasado mas de 2 veces en el pago de las cuotas

```python
# crear variable que incluya aquellos que no han realizado ningun pago en
# los ultimos noventa dias
dataset["deudores_mayor_90_dias"] = (dataset["next_pymnt_d_date"] - dataset["last_pymnt_d_date"]).dt.days >= 90
```

```python
#funcion mal pagador
def funcion_mal_pagador(dataset):
    if dataset["deudores_mayor_90_dias"] == True or dataset["loan_status"]=="Charged Off" or dataset["delinq_2yrs"]>=2:
        return True
    else:
        return False

dataset["mal_pagador"] = dataset.apply(funcion_mal_pagador,axis=1)
```

```python
# Nulos de emp_title y emp_length: SI no se tiene info, se completara con No especificado
def null_emp_title(emp_title):
    if pd.isnull(emp_title):
        return "No especificado"
    else:
        return emp_title

dataset["emp_title"] = dataset["emp_title"].apply(null_emp_title)

def null_emp_length(emp_length):
    if pd.isnull(emp_length):
        return "No especificado"
    else:
        return emp_length

dataset["emp_length"] = dataset["emp_length"].apply(null_emp_length)
```

```python
# Nulos annual_inc: Se rellena con la mediana
def null_annual_inc(annual_inc,mediana):
    if pd.isnull(annual_inc):
        return mediana
    else:
        return annual_inc

mediana = dataset["annual_inc"].median()
dataset["annual_inc"] = dataset["annual_inc"].apply(lambda x: null_annual_inc(x,mediana))
```

```python
# Nulos delinq_2yrs: Se rellena con ceros porque la mayoria de los nulos son aquellos
# que no han tenido retrasos
def null_delinq_2yrs(delinq_2yrs):
    if pd.isnull(delinq_2yrs):
        return 0
    else:
        return delinq_2yrs

dataset["delinq_2yrs"] = dataset["delinq_2yrs"].apply(lambda x: null_delinq_2yrs(x))
```

```python
# Nulos next_pymnt_d: los nulos son aquellos que ya fueron pagados o que fueron
# cancelados
def null_next_pymnt_d(dataset):
    if pd.isnull(dataset["next_pymnt_d"]) and dataset["loan_status"] == "Fully Paid":
        return "Finish"
    else:
        return dataset["next_pymnt_d"]

dataset["next_pymnt_d_prueba"] = dataset.apply(lambda x: null_next_pymnt_d(x),axis=1)


def null_next_pymnt_d(dataset):
    if pd.isnull(dataset["next_pymnt_d_prueba"]) and dataset["loan_status"] == "Charged Off":
        return "Canceled"
    else:
        return dataset["next_pymnt_d_prueba"]

dataset["next_pymnt_d_prueba"] = dataset.apply(lambda x: null_next_pymnt_d(x),axis=1)
```

```python
# mths_since_last_delinq
dataset["mths_since_last_delinq"].value_counts(dropna=False)
x=dataset[dataset["mths_since_last_delinq"].isna()]
x['delinq_2yrs'].value_counts()
# los nulos se concentran en los que no han tenido retraso por lo que para
# reemplazarlos se creara la categoria "no delinq" y los que si tuvieron mas
# de un retraso se completara con la moda
dataset["mths_since_last_delinq"] = dataset["mths_since_last_delinq"].apply(lambda x: "No_delinq" if pd.isnull(x) else x)

dataset["mths_since_last_delinq"].value_counts()
y=dataset[dataset["mths_since_last_delinq"]!= "No_delinq"]
moda_mths_dlinq=y["mths_since_last_delinq"].mode() #moda de los meses desde el ultimo retraso

dataset["mths_since_last_delinq"] = dataset["mths_since_last_delinq"].apply(lambda x: moda_mths_dlinq if pd.isnull(x) else x)

dataset["mths_since_last_delinq"].value_counts(dropna=False)
```

```python
# tot_hi_cred_lim
dataset["tot_hi_cred_lim"].value_counts(dropna=False)

x=dataset[dataset["tot_hi_cred_lim"].isna()]
x["settlement_status"].value_counts()

# no tienen techo maximo de monto de credito definido los que no aplican
dataset["tot_hi_cred_lim"] = dataset["tot_hi_cred_lim"].apply(lambda x: "No_Aplica" if pd.isnull(x) else x)
```

```python
# last_pymnt_d_date
dataset["last_pymnt_d_date"].value_counts(dropna=False)
x=dataset[dataset["last_pymnt_d_date"].isna()]
x['loan_status'].value_counts()

# los nulos en last_pymnt_d_date se concentran en los prestamos con status
# charged off por lo que se reemplazara esta variable

dataset["last_pymnt_d_date"] = dataset["last_pymnt_d_date"].apply(lambda x: "Charged_Off" if pd.isnull(x) else x)
dataset["last_pymnt_d_date"].value_counts(dropna=False)

dataset["last_pymnt_d2"] = dataset["last_pymnt_d"].apply(lambda x: "Charged_Off" if pd.isnull(x) else x)
```
